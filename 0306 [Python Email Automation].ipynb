```python
import requests
from bs4 import BeautifulSoup
from openpyxl import Workbook
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
from selenium.webdriver.common.keys import Keys
import pandas as pd 
data = pd.read_csv("C:\\Users\\Playdata\\Example.csv")


driver = webdriver.Chrome()
driver.implicitly_wait(3)

wait = WebDriverWait(driver, 3)

driver.get("http://www.google.com")

# sing-in 링크 클릭 
driver.find_element_by_css_selector('a.gb_2').click()
time.sleep(2)

# email주소 입력 
driver.find_element_by_name("identifier").send_keys("playdatamini")
driver.find_element_by_name("identifier").send_keys(Keys.ENTER)

# pwd 입력 
driver.find_element_by_name("password").send_keys("playdata1234")
driver.find_element_by_name("password").send_keys(Keys.ENTER)

### --- 로그인 완료 ---- ### 



# gmail 아이콘 클릭 
# driver.find_element_by_css_selector("a.gb_d").click()
# page 전환이기 때문에 css 를 못 찾을 수도 있으니 wait를 넣어보았다.
gmail_button = wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR, "a.gb_d")))
gmail_button.click()
time.sleep(2)





for i,v in enumerate(data):    
    # compose button 
    driver.find_element_by_css_selector("div.T-I.T-I-KE.L3").click()
    time.sleep(2)

    # recipient
    driver.find_element_by_name("to").send_keys(data.loc[i]["email"])
    driver.find_element_by_name("to").send_keys(Keys.ENTER)
    # 제목
    driver.find_element_by_name("subjectbox").send_keys("안녕하세요")
    # 내용
    body_tags = driver.find_element_by_css_selector("div.Am.Al.editable.LW-avf.tS-tW")
    body_contents = f"""안녕하세요! {data.loc[i]["name"]}님
    오늘 하루도 화이팅하시고 좋은 하루 보내시길 바라겠습니다 :)
    """
    body_tags.send_keys(body_contents)
    time.sleep(3)
    driver.find_element_by_css_selector("div.T-I.J-J5-Ji.aoO.v7.T-I-atl.L3").click()
    time.sleep(3)

driver.quit()
```
